```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Dataset_Banking_chatbot.csv", sep=',', encoding='latin1')
```

```python
df.head()
```

```python
import json
```

```python
data = []

for _, row in df.iterrows():
    data.append({
        "messages": [
            {
                "role": "user",
                "content": row["Query"]
            },
            {
                "role": "assistant",
                "content": row["Response"]
            }
        ]
    })

with open("banking_dataset.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```
